```python
import pandas as pd
import numpy as np

data = pd.Series([1, 2, 3, 4, 5])
print(data)
```

```python
#series with custom index
data = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
print(data)
```

```python
#accessing elements
print(data['a'])  # Access by index label
# print(data[0])    # Access by integer position
```

```python
#creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'Salary': [50000, 60000, 70000]
}

df = pd.DataFrame(data)
print(df)
```

```python
#accessing data in DataFrame
print(df['Name'])  # Access a single column
print(df.iloc[0])  # Access by integer position (row 0)
print(df.loc[1])   # Access by index label (row 0)
```

```python
#index objects in pandas
print(df.index)  # Access the index of the DataFrame
print(df.columns)  # Access the column labels of the DataFrame
```

```python
#re-indexing a DataFrame
new_df = df.reindex([2, 0, 1])
print(new_df)
```

```python
#drop columns
new_df = df.drop('Age', axis=1)
print(new_df)
```

```python
#row dropping
new_df = df.drop(1)
print(new_df)
```

```python
#selecting specfic columns
print(df[['Name', 'Salary']])
```

```python
#conditional selection
print(df[df['Age'] > 30])
```

```python
#DATA ALIGNMENT IN PANDAS
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([4, 5, 6], index=['b', 'c', 'd'])    
print(s1 + s2)
```

```python
#sorting
print(df.sort_values(by='Age'))  # Sort by Age
print("")
print(df.sort_values(by='Salary', ascending=False))  # Sort by Salary in descending order

```

```python
#rank
print(df['Salary'].rank())  # Rank the Salary column
# rank descending
print(df['Salary'].rank(ascending=False))  # Rank the Salary column in descending order
```
