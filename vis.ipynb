Part A: Basic Visualization (Matplotlib)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")

print(df.head())
print(df.columns)   # IMPORTANT check
```

```python
plt.figure()

plt.plot(df['math score'])

plt.title("Math Score Trend")
plt.xlabel("Student Index")
plt.ylabel("Math Score")

plt.xlim(0, 1000)     # axis limit
plt.ylim(0, 100)

plt.grid()

plt.show()
```

```python
plt.figure()

plt.scatter(df['math score'], df['reading score'])

plt.title("Math vs Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

plt.xlim(0, 100)
plt.ylim(0, 100)

plt.grid()

plt.show()
```

```python
plt.figure()

plt.hist(df['writing score'], bins=10)

plt.title("Writing Score Distribution")
plt.xlabel("Writing Score")
plt.ylabel("Frequency")

plt.xticks(rotation=45)   # tick rotation

plt.grid()

plt.show()
```

Part B: Subplots & Multiple Plots

```python
subjects = ['Math', 'Reading', 'Writing']

avg_scores = [
    df['math score'].mean(),
    df['reading score'].mean(),
    df['writing score'].mean()
]
```

```python
plt.figure(figsize=(10, 8))

# Line
plt.subplot(2, 2, 1)
plt.plot(df['math score'])
plt.title("Line Plot")
plt.grid()

# Scatter
plt.subplot(2, 2, 2)
plt.scatter(df['math score'], df['reading score'])
plt.title("Scatter Plot")
plt.grid()

# Bar
plt.subplot(2, 2, 3)
plt.bar(subjects, avg_scores)
plt.title("Bar Chart")

# Histogram
plt.subplot(2, 2, 4)
plt.hist(df['writing score'], bins=10)
plt.title("Histogram")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line
axs[0, 0].plot(df['math score'])
axs[0, 0].set_title("Line Plot")
axs[0, 0].grid()

# Scatter
axs[0, 1].scatter(df['math score'], df['reading score'])
axs[0, 1].set_title("Scatter Plot")
axs[0, 1].grid()

# Bar
axs[1, 0].bar(subjects, avg_scores)
axs[1, 0].set_title("Bar Chart")

# Histogram
axs[1, 1].hist(df['writing score'], bins=10)
axs[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

Part C: Advanced Customization

```python
plt.figure()

# Plot with label (for legend)
plt.scatter(df['math score'], df['reading score'], label="Students")

# Highlight important point (max math score)
max_idx = df['math score'].idxmax()

plt.scatter(df.loc[max_idx, 'math score'],
            df.loc[max_idx, 'reading score'],
            color='red', label="Top Math Score")

# Annotation
plt.annotate("Top Score",
             (df.loc[max_idx, 'math score'],
              df.loc[max_idx, 'reading score']),
             xytext=(80, 90),
             arrowprops=dict(arrowstyle="->"))

plt.title("Highlighted Scatter Plot")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure()

plt.hist(df['writing score'], bins=10)

plt.title("Writing Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.savefig("output.png")   # SAVE FILE

plt.show()
```

Part D: Seaborn Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")

print(df.head())
print(df.columns)
```

```python
sns.scatterplot(x='math score', y='reading score', hue='gender', data=df)

plt.title("Math vs Reading (Colored by Gender)")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

plt.show()
```

```python
sns.lineplot(x=df.index, y='writing score', data=df)

plt.title("Writing Score Trend")
plt.xlabel("Index")
plt.ylabel("Writing Score")

plt.show()
```

```python
sns.boxplot(x='gender', y='math score', data=df)

plt.title("Math Score Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Math Score")

plt.show()
```

```python
sns.pairplot(df[['math score', 'reading score', 'writing score']])

plt.show()
```

```python
g = sns.FacetGrid(df, col="gender")

g.map(plt.hist, "math score")

plt.show()
```

```python
sns.jointplot(x='math score', y='reading score', data=df, kind='scatter')

plt.show()
```

Math and reading scores show a positive correlation
Students with higher math scores tend to have higher reading scores
Box plot shows score distribution differences between genders
Pair plot confirms strong relationships among all three scores

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['math score'],
           df['reading score'],
           df['writing score'])

ax.set_title("3D Scatter Plot of Scores")
ax.set_xlabel("Math Score")
ax.set_ylabel("Reading Score")
ax.set_zlabel("Writing Score")

plt.show()
```

1. What patterns did you observe?
The visualizations show a strong positive relationship between math, reading, and writing scores. Students who score high in one subject generally score high in the others, indicating consistent academic performance. Most students have scores in the mid-to-high range, as seen in the histogram, with fewer students scoring very low. The overall distribution of scores is fairly balanced without extreme variations.
2. Which plot was most useful and why?
The scatter plot with hue (gender) was the most useful because it clearly shows the relationship between two variables (math and reading scores) while also distinguishing categories using color. This makes it easy to understand both correlation and group differences in a single plot, making it more informative compared to other plots.
3. Any outliers or trends?
There are a few mild outliers, mainly students with very low scores compared to the majority. However, no extreme anomalies are present. The main trend observed is a consistent positive correlation between all three subjects, indicating that students tend to perform similarly across math, reading, and writing.
