```python
import numpy as np
arr= np.array([1, 2, 3, 4, 5])
print("Array: ",arr)
```

```python
#shape, data type, size, and dimensions
print("Shape: ",arr.shape)
print("Data Type: ",arr.dtype)
print("Size: ",arr.size)
print("Dimensions: ",arr.ndim)
```

```python
#zero array
zero_arr = np.zeros((3, 4))
print("Zero Array: \n", zero_arr)

#ones array
ones_arr = np.ones((2, 5))
print("Ones Array: \n", ones_arr)

#random array
random_arr = np.random.rand(3, 3)
print("Random Array: \n", random_arr)

#identity matrix
identity_matrix = np.eye(4)
#identity_matrix = np.identity(5)
print("Identity Matrix: \n", identity_matrix)
```

```python
#item size
print("Item Size: ", arr.itemsize, "bytes")

#reshape
reshaped_arr = arr.reshape(5, 1)
print("Reshaped Array: \n", reshaped_arr)
```

```python
#slicing
sliced_arr = arr[1:4]
print("Sliced Array: ", sliced_arr)
```

```python
#line space
linspace_arr = np.linspace(0, 10, 5)
print("Linspace Array: ", linspace_arr)
```

```python
#max, min, sum, sqrt
print("Max: ", arr.max())
print("Min: ", arr.min())           
print("Sum: ", arr.sum())   
print("Square Root: ", np.sqrt(arr))
```

```python
#ravel
raveled_arr = arr.ravel()
print("Flattened Array: ", raveled_arr)
```

```python
#basic array operations
#join, split, search, sort, indexing, slicing, iteration, copy, array shape manipulation.
```

```python
#join - hstack
arr1 = np.array([[1, 2, 3],[4, 5, 6]])
arr2 = np.array([[7, 8, 9],[10, 11, 12]])
joined_arr = np.concatenate((arr1, arr2), axis=1)
print("Joined Array: ", joined_arr)

#join vstack
vstack_arr = np.vstack((arr1, arr2))
print("VStacked Array: \n", vstack_arr)

#join hstack    
hstack_arr = np.hstack((arr1, arr2))
print("HStacked Array: ", hstack_arr)
```

```python
#split
arr4 = np.array([[1, 2, 3, 4],[5, 6, 7, 8]])
split_arr = np.split(arr4, 2, axis=1)
print("Split Arrays:", split_arr)
```

```python
search_arr = np.array([10, 20, 30, 40, 50])
index = np.where(search_arr == 30)
print("Index of 30: ", index)
```

```python
#sort

arr3 = np.array([5, 2, 9, 1, 5, 6])
sorted_arr = np.sort(arr3)
print("Sorted Array: ", sorted_arr)
```

```python
#indexing
print("First Element: ", arr[0])
print("Last Element: ", arr[-1])

#slicing
sliced_arr = arr[1:4]
print("Sliced Array: ", sliced_arr)

#iterating
for element in arr:
    print("Element: ", element)
```

```python
#copy
copied_arr = arr.copy()
copied_arr[0] = 100
print("Copied Array: ", copied_arr) 
print("Original Array: ", arr)
```

```python
#array shape manipulation
reshaped_arr = arr.reshape(5, 1)
print("Reshaped Array: \n", reshaped_arr)
```
